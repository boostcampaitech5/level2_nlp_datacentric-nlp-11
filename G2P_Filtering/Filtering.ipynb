이제 모든 데이터에 대해서 부산대학교 맞춤법 검사기를 활용한 G2P Filtering을 진행합니다.

```python
# 모든 데이터에 대해서 함수 실행
from tqdm import tqdm

train_fix_spacing = pd.read_csv('train.csv')
pcs_of_wrong_words = []
for sample in tqdm(train.iterrows(), total = len(train.index), desc = 'Description', ncols = 100, ascii = ' ='):
    pcs = 0
    text = sample[1][1]
    response = requests.post('http://164.125.7.61/speller/results', data={'text1': text})
    data = response.text.split('data = [', 1)[-1].rsplit('];', 1)[0]
    try:
        data = json.loads(data)
        for info in data['errInfo']:
            # 띄어쓰기 오류라면 고쳐줍니다.
            if info['orgStr'].replace(" ", "") == info['candWord'].split("|")[0].replace(" ", ""):
                text = re.sub(info['orgStr'], info['candWord'].split("|")[0], text)
                train_fix_spacing.loc[sample[0], 'text'] = text
            # . 혹은 …. 변환의 경우 무시합니다.
            elif info['candWord'].split("|")[0][-1] == '.' or info['candWord'].split("|")[0][-1] == '….':
                pass
            # 기타 오류의 개수를 셉니다.
            else:
                pcs += 1
    except:
        pass
    # 오류의 개수를 저장합니다.
    pcs_of_wrong_words.append(pcs) 
print(pcs_of_wrong_words)
```

원본 데이터인 base_train을 가져온 후 오류의 개수(errors) 열을 추가해줍니다.

```python
base_train = train
base_train['errors'] = pcs_of_wrong_words
base_train.tail()
```

오류가 1개도 없는 데이터, 1개까지는 허용한 데이터, 2개까지는 허용한 데이터로 나누어 저장합니다.

```python
less_than_one_error_base = base_train[base_train['errors'] < 1]
less_than_two_errors_base = base_train[base_train['errors'] < 2]
less_than_three_errors_base = base_train[base_train['errors'] < 3]
```

```python
less_than_one_error_base.to_csv('less_than_one_error_base.csv', index=False)
less_than_two_errors_base.to_csv('less_than_two_errors_base.csv', index=False)
less_than_three_errors_base.to_csv('less_than_three_errors_base.csv', index=False)
base_train.to_csv('base_train.csv', index=False)
```
